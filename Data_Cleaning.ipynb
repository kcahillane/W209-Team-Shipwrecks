```python
PH_authors = pd.read_excel('A de Graauw 2014 - Geodatabase of Ancient Ports and Harbors (v 1-1).xlsx', 
                           'Ancient authors by date', 
                           dtype = object)
PH_authors = PH_authors[PH_authors['Author'].notna()]
PH_authors.reset_index(inplace = True)
PH_authors
```

```python
PH_places = pd.read_excel('A de Graauw 2014 - Geodatabase of Ancient Ports and Harbors (v 1-1).xlsx', 
                          'PLACES')
#PH_places = PH_places[PH_places['AUTH_ANC'].notna()]
#PH_places.reset_index(inplace = True)
PH_places = PH_places.drop(columns = ['AM', 'PP', 'RE', 'BW', 'QU', 'MO',
                                      'CN', 'SL', 'SH', 'PH', 'CO',
                                      'AUTH_MOD', 'NB', 'NAME_MOD'], axis = 1)
PH_places["begin"] = 1000
PH_places["end"] = -1000
authcount = -1
for author in PH_authors["Author"].unique():
    authcount+= 1
    for i in range(len(PH_places)):
        if author in str(PH_places.loc[i, "AUTH_ANC"]):
            if PH_places.loc[i, "begin"] > PH_authors.loc[authcount, "begin"]:
                PH_places.loc[i, "begin"] = PH_authors.loc[authcount, "begin"]
            if PH_places.loc[i, "end"] < PH_authors.loc[authcount, "end"]:
                PH_places.loc[i, "end"] = PH_authors.loc[authcount, "end"]  
PH_places.loc[PH_places["begin"] == 1000, "begin"] = None
PH_places.loc[PH_places["end"] == -1000, "end"] = None
PH_places
```

```python
PH_places.to_csv("ports_with_dates", index = False)
```

```python
SW = pd.read_excel('Michael McCormick et al., Summary Geodatabase of Shipwrecks AD 1-1500, Status 2008.xlsx',
                   'Shipwrecks 2008 Main',
                   dtype = object)
SW["amph detail"] = ""

for index, row in SW.iterrows():
    amph_detail = ""
    if row["Cargo 1"] == "amphoras":
        amph_detail = row["Type 1"]
    elif row["Cargo 2"] == "amphoras":
        amph_detail = row["Type 2"]
    elif row["Cargo 3"] == "amphoras":
        amph_detail = row["Type 3"]
    row["amph detail"] = amph_detail

for index, row in SW.iterrows():
    if "Dr" in str(row["amph detail"]):
        if "Dressel" not in str(row["amph detail"]):
            row["amph detail"] = str(row["amph detail"]).replace("Dr", "Dressel")
    if "Afr" in str(row["amph detail"]):
        if "Africana" not in str(row["amph detail"]):
            row["amph detail"] = str(row["amph detail"]).replace("Afr", "Africana")
    if "PE" in str(row["amph detail"]):
        row["amph detail"] = str(row["amph detail"]).replace("PE", "Pélichet")
    if "Pelichet" in str(row["amph detail"]):
        row["amph detail"] = str(row["amph detail"]).replace("Pelichet", "Pélichet")     
    if "LR" in str(row["amph detail"]):
        row["amph detail"] = str(row["amph detail"]).replace("LR", "Late Roman Amphora")  
    if "LRA" in str(row["amph detail"]):
        row["amph detail"] = str(row["amph detail"]).replace("LRA", "Late Roman Amphora")
    if "Beltran" in str(row["amph detail"]):
        row["amph detail"] = str(row["amph detail"]).replace("Beltran", "Beltrán")
    if "Lam" in str(row["amph detail"]):
        row["amph detail"] = str(row["amph detail"]).replace("Lam", "Lamboglia")
```
