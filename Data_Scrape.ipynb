```python
import requests
```

```python
table_of_contents  = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'K', 'L', 'M', 'O', 'P', 'R', 'S', 'T', 'U', 'V']
```

```python
#scraping all item URLs
soup_urls = []
for letter in table_of_contents:
    home_url = "https://archaeologydataservice.ac.uk/archives/view/amphora_ahrb_2005/cat_amph.cfm?l="+letter+"&CFID=7e135a74-2988-4da6-917c-63078164c7a7&CFTOKEN=0"
    home_page = requests.get(home_url)
    home_soup = BeautifulSoup(home_page.content, 'html.parser')
    home_list = home_soup.find_all('p', class_="inny")
    for item in home_list:
        url = item.find('a', class_ = "update")['href']
        soup_urls.append(url)
soup_urls
    
```

```python
len(soup_urls)
```

```python
#scraping amph locations
df_loc = []
for url in soup_urls:
    item_url = "https://archaeologydataservice.ac.uk/archives/view/amphora_ahrb_2005/"+url
    item_page = requests.get(item_url)
    item_soup = BeautifulSoup(item_page.content, 'html.parser')
    Name = item_soup.find('h2', class_ = "left").text
    item_sections = item_soup.find_all('td', class_ = "dets")
    loc_list = []
    for item in item_sections:
        if "href" in str(item):
            origins = item.find_all('a')
            if "pplace" in str(origins):
                locations = item.find_all('a')
                for loc in locations:
                    loc_list.append(str(loc.text)[4:-1])
    df_loc.append(loc_list)
    
print(df_loc)
```

```python
#scraping amph contents
df_contents = []
for url in soup_urls:
    item_url = "https://archaeologydataservice.ac.uk/archives/view/amphora_ahrb_2005/"+url
    item_page = requests.get(item_url)
    item_soup = BeautifulSoup(item_page.content, 'html.parser')
    Name = item_soup.find('h2', class_ = "left").text
    item_sections = item_soup.find_all('td', class_ = "dets")
    con_list = []
    for item in item_sections:
        if "href" in str(item):
            origins = item.find_all('a')
            if "conts" in str(origins):
                contents = item.find_all('a')
                for content in contents:
                    con_list.append(str(content.text)[4:-1])
    df_contents.append(con_list)
    
print(df_contents)
```

```python
#scraping classification
df_class = []
for url in soup_urls:
    item_url = "https://archaeologydataservice.ac.uk/archives/view/amphora_ahrb_2005/"+url
    item_page = requests.get(item_url)
    item_soup = BeautifulSoup(item_page.content, 'html.parser')
    Name = item_soup.find('h2', class_ = "left").text
    item_sections = item_soup.find_all('td', class_ = "dets")
    class_list = []
    for item in item_sections:
        if "Classification" in str(item):
            classes = str(item.text).splitlines(False)[1:]
            for class_ in classes:
                class_list.append(class_[4:])
    df_class.append(class_list)
df_class
                

    
```

```python
#scraping Name
df_name = []
for url in soup_urls:
    item_url = "https://archaeologydataservice.ac.uk/archives/view/amphora_ahrb_2005/"+url
    item_page = requests.get(item_url)
    item_soup = BeautifulSoup(item_page.content, 'html.parser')
    Name = item_soup.find('h2', class_ = "left").text
    item_sections = item_soup.find_all('td', class_ = "dets")
    df_name.append(Name)
print(df_name)
```

```python
df_amph = pd.DataFrame(list(zip(df_name, df_loc, df_contents, df_class)),
               columns =['Name', 'Origin', 'Contents', 'Classifications'])
```
